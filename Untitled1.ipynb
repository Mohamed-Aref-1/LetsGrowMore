# Building to model

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
le = le = LabelEncoder()
```

```python
data_frame['class'] = le.fit_transform = (data_frame['class'])
data_frame.head(100)
```

```python
x = data_frame.drop(columns = 'class' , axis = 1)
y = data_frame['class']
```

```python
print(f'The shape of Y is {y.shape}, and x shape is {x.shape}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state= 2, test_size = 0.17)
```

```python
print(f'X_train shape is {x_train.shape}, and y_train shape is {y_train.shape}')
```

# Logistic regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , ConfusionMatrixDisplay
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
logistic_model= LogisticRegression(max_iter=1000)
```

```python
logistic_model.fit(x_train , y_train)
```

```python
y_predict =logistic_model.predict(x_test)
score_lr=accuracy_score(y_test,y_predict)
```

```python
def report(model):
    preds=model.predict(x_test)
    print(classification_report(preds,y_test))
    cm = confusion_matrix(y_test, preds, labels = model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = model.classes_)
    disp.plot()
#     plt.show()
```

```python
print('Logistic Regression')
report(logistic_model)
print(f'Accuracy: {round(score_lr*100,2)}%')
plt.grid(False)
```

```python
#Accuracy of the Logistic regression :  96.67%
```

# KNeighborsClassifier


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors= 10)
```

```python
KNN.fit(x_train, y_train)
```

```python
y_prediction_knn = KNN.predict(x_test)
score_Knn = accuracy_score(y_prediction_knn, y_test)
```

```python
print('K Nearest Neighbours Algorithm')
report(KNN)
print(f'Accuracy: {round(score_Knn*100,2)}%')
plt.grid(False)
```

```python
#Accuracy of the K Nearest Neighbours :  92.31%
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decesion_tree = DecisionTreeClassifier().fit(x_train , y_train)
```

```python
y_pred_dt = decesion_tree.predict(x_test)
score_Dt = accuracy_score(y_pred_dt ,  y_test)
```

```python
print('Decision Tree Classifier Algorithm')
report(decesion_tree)
print(f'Accuracy: {round(score_Dt*100,2)}%')
```

```python
#Accuracy of Decision Tree :  96.15%
```
